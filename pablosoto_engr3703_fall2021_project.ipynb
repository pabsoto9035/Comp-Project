# Unforced

Overdamped System

```python
%matplotlib inline
from math import*
import numpy as np
import matplotlib.pyplot as plt

#Classical Method #'s
c1=1./6
c2=2./6
c3=2./6
c4=1./6

a2=1./2
a3=1./2
a4=1.

b21=1./2
b31=0.
b32=1./2
b41=0.
b42=0.
b43=1.

zeta=2
omega=1

Fm=lambda t:0
f_x=lambda t,x,v:v
f_v=lambda t,x,v:Fm(t)-2*zeta*omega*v-omega**2*x

def rk4_x(ti,xi,vi,dt):
    K1=f_x(ti,xi,vi)
    K2=f_x(ti+a2*dt,xi+b21*K1*dt,vi)
    K3=f_x(ti+a3*dt,xi+b31*K1*dt+b32*K2*dt,vi)
    K4=f_x(ti+a4*dt,xi+b41*K1*dt+b42*K2*dt+b43*K3*dt,vi)
    xip1=xi+(c1*K1+c2*K2+c3*K3+c4*K4)*dt
    return xip1

def rk4_v(ti,xi,vi,dt):
    K1=f_v(ti,xi,vi)
    K2=f_v(ti+a2*dt,xi,vi+b21*K1*dt)
    K3=f_v(ti+a3*dt,xi,vi+b31*K1*dt+b32*K2*dt)
    K4=f_v(ti+a4*dt,xi,vi+b41*K1*dt+b42*K2*dt+b43*K3*dt)
    vip1=vi+(c1*K1+c2*K2+c3*K3+c4*K4)*dt
    return vip1

ti=0.0
tf=100.0
n=5000
dt=(tf-ti)/n
t=np.zeros(n+1)
x=np.zeros(n+1)
v=np.zeros(n+1)
x_iv=0.0
v_iv=0.2
t[0]=ti
x[0]=x_iv
v[0]=v_iv

for i in range (1,n+1):
    t[i]=ti+i*dt
    x[i]=rk4_x(t[i-1],x[i-1],v[i-1],dt)
    v[i]=rk4_v(t[i-1],x[i-1],v[i-1],dt)
    
fig=plt.figure(figsize=(8,9))
ax1=plt.subplot(221)
plt.title("Displacement for an Overdamped System")
plt.plot(t,x, label="x vs. t", color="b",linewidth="2.0")
plt.xlabel("t(s)")
plt.ylabel("v(m/s)")

ax2=plt.subplot(212,sharex=ax1)
plt.title("Velocity for an Overdamped System")
plt.plot(t,v, label="v vs. t", color="b",linewidth="2.0")
plt.xlabel("t(s)")
plt.ylabel("v(m/s)")

plt.tight_layout()
plt.show()

```

# Unforced Analysis

When going through the unforced analysis, all of the graphs reached equilibrium. Even though the start of graph was different they all, eventually, ended up going in a straigh horizontal line. The overdamped and critically damped experiments showed a very similar graph shape, however, the critically damped line reached equilibrium much quicker than the overdamped. The underdamped graph trend was the most unique out of the three. The frequency between each oscillation was also the highest. However, as it reached closer to equilibrium, the oscillations decreased. All three graphs reached equilibrium at the same point.

# Forced Analysis

When using different equations in the forced analysis, the oscillation can change dramatically throughout the graphs. For example, linear and constant graphs will have very frequent oscilations where as the quadratic or cubic equations will not. Based on the results of the graphs, as the value of the exponent, or the higher the order, the oscillation will be less frequent. Also, as the order of the function increased, the graph had a tougher time to reach equillibrium. When using the osilating equation provided, equilibirium was even further from being reached.
